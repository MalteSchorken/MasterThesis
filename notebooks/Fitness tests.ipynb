```python
from evoaudio.base_sample import BaseSample
from evoaudio.individual import BaseIndividual
from evoaudio.population import Population
from evoaudio.sample_library import SampleLibrary
from evoaudio.fitness import multi_onset_fitness_cached
from evoaudio.pitch import Pitch
from evoaudio.target import Target
```

```python
sample_lib = SampleLibrary()
```

```python
instruments = ["Piano", "Piano", "Piano"]
styles = ["The_Grandeur", "The_Grandeur", "The_Grandeur"]
pitches_C = [Pitch.c4, Pitch.e4, Pitch.g4]

individual_C = BaseIndividual.create_individual(sample_lib=sample_lib,
                                                n_samples=3, instruments=instruments,
                                                styles=styles, pitches=pitches_C)
```

```python
instruments = ["EgyptianFiddle", "Bass", "Balalaika", "Dilruba"]
styles = ["egyptian_fiddle", "Basses_SessionStringsPro", "balalaika", "dilruba"]
pitches = [Pitch.c4, Pitch.dis1, Pitch.e5, Pitch.ais4]

individual_1 = BaseIndividual.create_individual(sample_lib=sample_lib,
                                                n_samples=4, instruments=instruments,
                                                styles=styles, pitches=pitches)
```

```python
instruments = ["Bass", "Dilruba", "EgyptianFiddle", "Balalaika"]
styles = ["Basses_SessionStringsPro", "dilruba", "egyptian_fiddle", "balalaika"]
pitches = [Pitch.dis1, Pitch.ais4, Pitch.c4, Pitch.e5]

individual_2 = BaseIndividual.create_individual(sample_lib=sample_lib,
                                                n_samples=4, instruments=instruments,
                                                styles=styles, pitches=pitches)
```

```python
individual_C.recalc_fitness = True
objectives = [("stft", "js_divergence")]
target = Target(individual_1.to_mixdown(), [0], objectives)
multi_onset_fitness_cached(target, individual_2, objectives = objectives)
```

```python
individual_C.recalc_fitness = True
objectives = [("stft", "js_divergence"), ("pseudo_cqt", "js_divergence")]
target = Target(individual_1.to_mixdown(), [0], objectives)
multi_onset_fitness_cached(target, individual_1, objectives = objectives)
```
